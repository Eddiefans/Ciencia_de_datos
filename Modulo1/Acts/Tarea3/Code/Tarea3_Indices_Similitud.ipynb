#### 1. Del archivo con información de la contaminación de zona metropolitana (“contaminacion_zm.xlsx”), escoja las mediciones de al menos 4 estaciones que miden la calidad del aire (ej: Atemajac, Aguilas, Centro, Oblatos).

```python
import pandas as pd 
from CDIN import CDIN as cd

estaciones = ["Atemajac", "Miravalle", "Tlaquepaque", "Vallarta"]
excel = pd.ExcelFile("../Data/contaminacion_zm.xlsx")

df  = {}
for i in excel.sheet_names:
    if i in estaciones:
        df[i] = excel.parse(i)

df["Atemajac"].head()
```

Obtener un reporte de la calidad de los datos de al menos una tabla, para determinar la información básica de la base de datos.

```python
cd.dqr(df["Atemajac"])
```

#### 3. Con la ayuda del cálculo de las distancias, determine cuales estaciones tienen un comportamiento similar durante todo el año. La comparación será entre contaminantes del mismo tipo, por cada estación, es decir, se requiere comparar el “CO” de todas las estaciones, después se obtiene una nueva distancia entre “NO2” de cada estación y así sucesivamente.

```python
el = ["CO", "NO2", "O3", "PM10", "SO2"]
for i, v in df.items(): 
    
    for j in v.columns: 
        if j not in el:
            df[i] = df[i].drop(j, axis='columns')
            
    for j in el:
        df[i][j].fillna(df[i][j].mean(), inplace=True)

data = {}
for i in el:
    data[i] = pd.DataFrame(columns=[i + "_A", i + "_M", i + "_T", i + "_V"])
    for j in range(len(data[i].columns)):
        data[i][data[i].columns[j]] = df[estaciones[j]][i]
    data[i] = data[i].T

cd.dqr(df["Atemajac"])
```

#### 4. Realice una tabla con las medidas de similitud o distancia en cada una de las comparaciones.



```python
cd.pdistance_matrix(data["CO"],"braycurtis")
```

```python
cd.pdistance_matrix(data["NO2"],"braycurtis")
```

```python
cd.pdistance_matrix(data["O3"],"braycurtis")
```

```python
cd.pdistance_matrix(data["PM10"],"braycurtis")
```

```python
cd.pdistance_matrix(data["SO2"],"braycurtis")
```

```python
for sub in el:
    bray = cd.pdistance_matrix(data[sub],"braycurtis")
    similitud = 1
    for i in bray:
        for j in range(len(bray[i])):
            if bray[i][j] < similitud and bray[i][j] != 0:
                similitud = bray[i][j]
                s1 = i 
                s2 = j
    print("{} y {} son las estaciones más similares con {}% con respecto a {}".format(estaciones[s1], estaciones[s2], round(similitud*100, 2), sub))
```

#### 5.- Incluya en este apartado el índice de similitud que se haya encontrado (vea el archivo pdf, donde encontará las medidas de similitud más usadas en el análisis de datos ), su fórmula y un pequeño ejemplo de cómo se aplica en una serie de datos de prueba.

- Indice: Métrica Bray-Curtis
- Fórmula: $\frac{ \sum_{t=1}^n |x_{it} - x_{jt}|}{ \sum_{t=1}^n (x_{it} + x_{jt})}$
- Ejemplo: 

    Muestra 1: 1, 1, 0, 1

    Muestra 2: 1, 0, 1, 0

    $\sum_{t=1}^n |x_{it} - x_{jt}|$ = |1-1| + |1-0| + |0-1| + |1-0| = 3

    $\sum_{t=1}^n (x_{it} + x_{jt})$ = (1+1) + (1+0) + (0+1) + (1+0) = 5
    
    Similitud = 3/5 = 60%
